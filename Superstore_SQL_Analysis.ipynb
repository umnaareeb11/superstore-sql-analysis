```python
import sqlite3        
import pandas as pd   
import warnings
warnings.filterwarnings('ignore')

print("✅ Libraries imported successfully!")
print(f"  pandas version : {pd.__version__}")
print(f"  sqlite3 version: {sqlite3.sqlite_version}")
```

```python
conn = sqlite3.connect('SuperstoreDB.db')   
cursor = conn.cursor()

print("✅ Database 'SuperstoreDB' created successfully!")
```

```python
create_table_sql = """
-- MySQL / SQLite compatible table creation
CREATE TABLE IF NOT EXISTS orders (
    Row_ID       INTEGER,
    Order_ID     VARCHAR(25),
    Order_Date   TEXT,          -- DATE in MySQL
    Ship_Date    TEXT,          -- DATE in MySQL
    Ship_Mode    VARCHAR(50),
    Customer_ID  VARCHAR(20),
    Customer_Name VARCHAR(100),
    Segment      VARCHAR(50),
    Country      VARCHAR(100),
    City         VARCHAR(100),
    State        VARCHAR(100),
    Postal_Code  VARCHAR(20),
    Region       VARCHAR(50),
    Product_ID   VARCHAR(25),
    Category     VARCHAR(50),
    Sub_Category VARCHAR(50),
    Product_Name VARCHAR(300),
    Sales        DECIMAL(10,4),
    Quantity     INTEGER,
    Discount     DECIMAL(5,4),
    Profit       DECIMAL(10,4)
);
"""

cursor.execute(create_table_sql)
conn.commit()
print("✅ Table 'orders' created successfully inside SuperstoreDB!")
print()
print("MySQL Equivalent:")
print("""
CREATE DATABASE SuperstoreDB;
USE SuperstoreDB;

CREATE TABLE orders (
    Row_ID        INT,
    Order_ID      VARCHAR(25),
    Order_Date    DATE,
    Ship_Date     DATE,
    Ship_Mode     VARCHAR(50),
    Customer_ID   VARCHAR(20),
    Customer_Name VARCHAR(100),
    Segment       VARCHAR(50),
    Country       VARCHAR(100),
    City          VARCHAR(100),
    State         VARCHAR(100),
    Postal_Code   VARCHAR(20),
    Region        VARCHAR(50),
    Product_ID    VARCHAR(25),
    Category      VARCHAR(50),
    Sub_Category  VARCHAR(50),
    Product_Name  VARCHAR(300),
    Sales         DECIMAL(10,4),
    Quantity      INT,
    Discount      DECIMAL(5,4),
    Profit        DECIMAL(10,4)
);
""")
```

```python
df = pd.read_csv('Sample - Superstore.csv', encoding='latin-1')
df.columns = [
    'Row_ID', 'Order_ID', 'Order_Date', 'Ship_Date', 'Ship_Mode',
    'Customer_ID', 'Customer_Name', 'Segment', 'Country', 'City',
    'State', 'Postal_Code', 'Region', 'Product_ID', 'Category',
    'Sub_Category', 'Product_Name', 'Sales', 'Quantity', 'Discount', 'Profit'
]

df.to_sql('orders', conn, if_exists='replace', index=False)
conn.commit()

print(f"✅ Data imported successfully!")
print(f"   Total rows loaded : {len(df):,}")
print(f"   Total columns     : {len(df.columns)}")
print(f"   Columns           : {list(df.columns)}")
```

```python
def run_query(sql, title=''):
    """Execute a SQL query and return results as a formatted DataFrame."""
    if title:
        print(f"\n{'='*60}")
        print(f"  {title}")
        print(f"{'='*60}")
    result = pd.read_sql_query(sql, conn)
    print(result.to_string(index=False))
    return result

print("✅ Helper function ready!")
```

```python
sql_q1 = """
-- View the first 10 rows of the dataset
-- MySQL: SELECT * FROM orders LIMIT 10;
SELECT *
FROM   orders
LIMIT  10;
"""

print("Query 1: First 10 Rows")
print("-" * 40)
result = pd.read_sql_query(sql_q1, conn)
print(result.to_string(index=False))
```

```python
sql_q2 = """
-- Count total number of records in the dataset
SELECT COUNT(*) AS Total_Records
FROM   orders;
"""

run_query(sql_q2, 'Query 2: Total Record Count')
```

```python
sql_q4 = """
-- Check NULL values in each important column
SELECT
    SUM(CASE WHEN Order_ID      IS NULL THEN 1 ELSE 0 END) AS Null_Order_ID,
    SUM(CASE WHEN Customer_Name IS NULL THEN 1 ELSE 0 END) AS Null_Customer,
    SUM(CASE WHEN Region        IS NULL THEN 1 ELSE 0 END) AS Null_Region,
    SUM(CASE WHEN Sales         IS NULL THEN 1 ELSE 0 END) AS Null_Sales,
    SUM(CASE WHEN Profit        IS NULL THEN 1 ELSE 0 END) AS Null_Profit,
    SUM(CASE WHEN Quantity      IS NULL THEN 1 ELSE 0 END) AS Null_Quantity,
    SUM(CASE WHEN Postal_Code   IS NULL THEN 1 ELSE 0 END) AS Null_Postal_Code
FROM orders;
"""

run_query(sql_q4, 'Query 4: NULL Value Check')
```

```python
sql_q5 = """
-- Count distinct values in key categorical columns
SELECT
    COUNT(DISTINCT Region)       AS Unique_Regions,
    COUNT(DISTINCT Category)     AS Unique_Categories,
    COUNT(DISTINCT Sub_Category) AS Unique_SubCategories,
    COUNT(DISTINCT Ship_Mode)    AS Unique_ShipModes,
    COUNT(DISTINCT Segment)      AS Unique_Segments,
    COUNT(DISTINCT State)        AS Unique_States,
    COUNT(DISTINCT Customer_ID)  AS Unique_Customers
FROM orders;
"""

run_query(sql_q5, 'Query 5: Distinct Values per Column')
```

```python
sql_q7 = """
-- Filter orders from the Technology category only
SELECT Order_ID, Product_Name, Category, Sales, Profit
FROM   orders
WHERE  Category = 'Technology'
LIMIT  10;
"""

run_query(sql_q7, 'Query 7: Technology Category Orders (first 10)')
print("\n💡 Insight: Category filtering lets buyers focus on their product line's performance.")
```

```python
sql_q8 = """
-- Show orders with sales amount exceeding $500
SELECT Order_ID, Customer_Name, Sales, Profit, Region
FROM   orders
WHERE  Sales > 500
ORDER  BY Sales DESC
LIMIT  10;
"""

run_query(sql_q8, 'Query 8: Orders with Sales > $500 (Top 10)')
print("\n💡 Insight: High-value orders deserve priority handling and follow-up.")
```

```python
sql_q9 = """
-- Filter orders placed between 1st Jan 2016 and 31st Dec 2016
SELECT Order_ID, Order_Date, Customer_Name, Sales
FROM   orders
WHERE  Order_Date BETWEEN '2016-01-01' AND '2016-12-31'
ORDER  BY Order_Date
LIMIT  10;
"""

run_query(sql_q9, 'Query 9: Orders Placed in Year 2016 (first 10)')
print("\n💡 Insight: Date range filters are essential for year-over-year comparisons.")
```

```python
sql_q10 = """
-- Orders from West region in Technology category with Sales > $200
SELECT Order_ID, Customer_Name, Region, Category, Sales, Profit
FROM   orders
WHERE  Region = 'West'
  AND  Category = 'Technology'
  AND  Sales > 200
LIMIT  10;
"""

run_query(sql_q10, 'Query 10: West + Technology + Sales > $200 (AND condition)')
print("\n💡 Insight: AND lets us narrow down to a very specific subset of records.")
```

```python
sql_q11 = """
-- Orders from East OR West region using OR operator
-- Equivalent using IN: WHERE Region IN ('East', 'West')
SELECT Order_ID, Region, State, Sales
FROM   orders
WHERE  Region IN ('East', 'West')
ORDER  BY Region, Sales DESC
LIMIT  10;
"""

run_query(sql_q11, 'Query 11: Orders from East or West (using IN)')
print("\n💡 Insight: IN is cleaner than writing multiple OR conditions.")
```

```python
sql_q12 = """
-- Products NOT in Furniture, with 'Phone' in their name
SELECT Product_Name, Category, Sub_Category, Sales, Profit
FROM   orders
WHERE  NOT Category = 'Furniture'
  AND  Product_Name LIKE '%Phone%'
LIMIT  10;
"""

run_query(sql_q12, 'Query 12: Non-Furniture Products with "Phone" in Name')
print("\n💡 Insight: LIKE with wildcards (%) helps search text without knowing the exact value.")
```

```python
sql_q13 = """
-- Calculate total sales for each region
SELECT
    Region,
    ROUND(SUM(Sales), 2)   AS Total_Sales,
    COUNT(*)               AS Total_Orders
FROM   orders
GROUP  BY Region
ORDER  BY Total_Sales DESC;
"""

run_query(sql_q13, 'Query 13: Total Sales by Region')
print("\n💡 Insight: The West region typically leads in total sales, guiding where to invest in marketing.")
```

```python
sql_q14 = """
-- Calculate total profit for each product category
SELECT
    Category,
    ROUND(SUM(Profit), 2)  AS Total_Profit,
    ROUND(SUM(Sales),  2)  AS Total_Sales,
    ROUND(SUM(Profit)/SUM(Sales)*100, 2) AS Profit_Margin_Pct
FROM   orders
GROUP  BY Category
ORDER  BY Total_Profit DESC;
"""

run_query(sql_q14, 'Query 14: Total Profit by Category')
print("\n💡 Insight: Technology has the highest profit margins — Furniture often runs at a loss due to discounts.")
```

```python
sql_q15 = """
-- Average sales value per unique customer
SELECT
    Customer_ID,
    Customer_Name,
    ROUND(AVG(Sales), 2) AS Avg_Sales_Per_Order,
    COUNT(*)             AS Total_Orders
FROM   orders
GROUP  BY Customer_ID, Customer_Name
ORDER  BY Avg_Sales_Per_Order DESC
LIMIT  10;
"""

run_query(sql_q15, 'Query 15: Top 10 Customers by Average Sales per Order')
print("\n💡 Insight: High-average customers are likely B2B buyers — worth nurturing for loyalty programs.")
```

```python
sql_q16 = """
-- Total quantity sold in each product category
SELECT
    Category,
    SUM(Quantity) AS Total_Units_Sold
FROM   orders
GROUP  BY Category
ORDER  BY Total_Units_Sold DESC;
"""

run_query(sql_q16, 'Query 16: Total Quantity Sold by Category')
print("\n💡 Insight: Office Supplies sell in the highest volume, even if each unit is cheap.")
```

```python
sql_q17 = """
-- Count of orders for each shipping method
SELECT
    Ship_Mode,
    COUNT(*)              AS Total_Orders,
    ROUND(SUM(Sales), 2)  AS Total_Sales
FROM   orders
GROUP  BY Ship_Mode
ORDER  BY Total_Orders DESC;
"""

run_query(sql_q17, 'Query 17: Orders by Ship Mode')
print("\n💡 Insight: Standard Class is the most popular shipping — customers balance cost vs. speed.")
```

```python
sql_q18 = """
-- Monthly sales summary across all years
SELECT
    STRFTIME('%Y-%m', Order_Date)  AS Year_Month,
    COUNT(*)                       AS Total_Orders,
    ROUND(SUM(Sales), 2)           AS Monthly_Sales,
    ROUND(SUM(Profit), 2)          AS Monthly_Profit
FROM   orders
GROUP  BY STRFTIME('%Y-%m', Order_Date)
ORDER  BY Year_Month
LIMIT  24;
"""

print("MySQL Equivalent uses: DATE_FORMAT(Order_Date, '%Y-%m') AS Year_Month")
print()
run_query(sql_q18, 'Query 18: Monthly Sales Summary (first 24 months)')
print("\n💡 Insight: Sales spike in November–December (holiday season) every year.")
```

```python
sql_biz7 = """
-- Shipping mode frequency and contribution to revenue
SELECT
    Ship_Mode,
    COUNT(*)                               AS Order_Count,
    ROUND(COUNT(*) * 100.0 /
          (SELECT COUNT(*) FROM orders), 2) AS Pct_of_Total,
    ROUND(SUM(Sales), 2)                   AS Total_Sales
FROM   orders
GROUP  BY Ship_Mode
ORDER  BY Order_Count DESC;
"""

run_query(sql_biz7, 'Business Q7: Most Frequently Used Ship Mode')
print("\n💡 Business Insight: Standard Class handles ~60% of orders — optimise logistics for this mode.")
```

```python
sql_biz9 = """
-- Check for duplicate Order_ID + Product_ID combinations
SELECT
    Order_ID,
    Product_ID,
    COUNT(*) AS Occurrences
FROM   orders
GROUP  BY Order_ID, Product_ID
HAVING COUNT(*) > 1
ORDER  BY Occurrences DESC
LIMIT  10;
"""

result = pd.read_sql_query(sql_biz9, conn)
print("Business Q9: Duplicate Record Check")
print("=" * 60)
if result.empty:
    print("✅ No duplicate Order_ID + Product_ID combinations found.")
else:
    print(f"⚠️  Found {len(result)} duplicate combinations:")
    print(result.to_string(index=False))
print("\n💡 Business Insight: Clean data = trustworthy analysis. Always validate before reporting.")
```

```python
sql_biz10 = """
-- Products below-average in sales but above-average in profit margin
SELECT
    Product_Name,
    Category,
    ROUND(SUM(Sales),  2)                        AS Total_Sales,
    ROUND(SUM(Profit), 2)                        AS Total_Profit,
    ROUND(SUM(Profit) / SUM(Sales) * 100, 2)    AS Profit_Margin_Pct
FROM   orders
GROUP  BY Product_Name, Category
HAVING
    SUM(Sales)  < (SELECT AVG(total_s) FROM
                      (SELECT SUM(Sales) AS total_s FROM orders GROUP BY Product_Name) t)
  AND
    SUM(Profit) / SUM(Sales) > 0.20
ORDER  BY Profit_Margin_Pct DESC
LIMIT  10;
"""

run_query(sql_biz10, 'Business Q10: Low Sales but High Profit Margin Products')
print("\n💡 Business Insight: These are 'hidden gems' — boost their visibility through promotions to maximise profit.")
```

```python
sql_q23 = """
-- Count total rows and distinct customers in the dataset
SELECT
    COUNT(*)                    AS Total_Rows,
    COUNT(DISTINCT Customer_ID) AS Distinct_Customers,
    COUNT(DISTINCT Order_ID)    AS Distinct_Orders,
    COUNT(DISTINCT Product_ID)  AS Distinct_Products
FROM orders;
"""

run_query(sql_q23, 'Query 23: Dataset Size Summary')
```

```python
sql_q24 = """
-- Count and total loss from negative-profit orders
SELECT
    COUNT(*)              AS Loss_Making_Orders,
    ROUND(SUM(Profit), 2) AS Total_Loss,
    ROUND(MIN(Profit), 2) AS Worst_Single_Loss
FROM   orders
WHERE  Profit < 0;
"""

run_query(sql_q24, 'Query 24: Loss-Making Orders')
print("\n💡 Insight: A significant number of orders result in losses, often due to high discounts.")
```

```python
sql_q25 = """
-- Check for data anomalies: zero/negative quantity or sales
SELECT
    SUM(CASE WHEN Quantity <= 0 THEN 1 ELSE 0 END) AS Invalid_Quantity_Count,
    SUM(CASE WHEN Sales    <= 0 THEN 1 ELSE 0 END) AS Zero_Or_Neg_Sales_Count,
    SUM(CASE WHEN Discount  > 1 THEN 1 ELSE 0 END) AS Discount_Over_100pct
FROM orders;
"""

run_query(sql_q25, 'Query 25: Data Anomaly / Consistency Check')
print("\n💡 Insight: A clean check here confirms the dataset is reliable for analysis.")
```

```python
sql_q26 = """
-- Profitability by discount band
SELECT
    CASE
        WHEN Discount = 0        THEN '0% - No Discount'
        WHEN Discount <= 0.10    THEN '1–10% Discount'
        WHEN Discount <= 0.20    THEN '11–20% Discount'
        WHEN Discount <= 0.40    THEN '21–40% Discount'
        ELSE                          '40%+ Discount'
    END                              AS Discount_Band,
    COUNT(*)                         AS Orders,
    ROUND(AVG(Profit), 2)            AS Avg_Profit,
    ROUND(SUM(Sales),  2)            AS Total_Sales
FROM   orders
GROUP  BY Discount_Band
ORDER  BY Avg_Profit DESC;
"""

run_query(sql_q26, 'Query 26: Profitability by Discount Band')
print("\n💡 Insight: Orders with 40%+ discounts often result in losses — reconsider deep discounting policies.")
```

```python
sql_q27 = """
-- Total profit and sales by customer segment
SELECT
    Segment,
    COUNT(DISTINCT Customer_ID)                  AS Customers,
    ROUND(SUM(Sales),  2)                        AS Total_Sales,
    ROUND(SUM(Profit), 2)                        AS Total_Profit,
    ROUND(SUM(Profit) / SUM(Sales) * 100, 2)    AS Profit_Margin_Pct
FROM   orders
GROUP  BY Segment
ORDER  BY Total_Profit DESC;
"""

run_query(sql_q27, 'Query 27: Profit by Customer Segment')
print("\n💡 Insight: Consumer segment has the most customers but Corporate tends to have better margins.")
```
